```python
#import the libraries
import numpy as np
import matplotlib.pyplot as mb
import pandas as pd
from apyori import apriori
```

```python
# import the dataset
store_data = pd.read_csv('https://raw.githubusercontent.com/yashy1626/ds_dataset/refs/heads/main/store_data.csv',header=None)
store_data
```

```python
# Data Preprocessing
# The Apriori library we are going to use requires our datasets to be in the form
# of a list to lists, where the whole datasets is a big list and ech transaction
# in the dataset is an inner list within tge outer big list
# currently we have in the form of a pandas dataframe

records=[]
for i in range(0,7501):        # i--> 1
  records.append([str(store_data.values[i,j]) for j in range(0,20)])

#i= take rows, j= will iterate column by column
```

```python
for i,j in enumerate(records):
  while 'nan' in records[i]:
    records[i].remove('nan')
```

```python
association_rules=apriori(records,
        min_support=0.0045,    #take the 35/7500 (why, )
        min_confidence=0.2,
        min_lift=3,
        min_lenght=2)      # 2 means we want at least two product in our rules
# min_confidence is a user defined threshold, chosen by the analyst
# why choose 20%
# a higher threshold (like 50%) might remove mant useful rules
# a lower threshold (like 5%) might generate too many weak rules

#min_lift=3
# user defined threshold
# we only want rules where A and B are at least 3x more likely to occur
# min_lift=3 --> to ensure strong association filter or strong corelation

association_results = list(association_rules)
```
