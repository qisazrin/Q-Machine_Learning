```python
import pyfpgrowth

# Step 1: Create transactions
transactions = [
 ['Milk', 'Bread', 'Butter'],
 ['Bread', 'Butter'],
 ['Milk', 'Egg'],
 ['Bread', 'Egg'],
 ['Bread', 'Butter', 'Egg', 'Milk']
]

print(transactions)
```

```python
# step 2 : FP_Growth frequent patterns
frequent=pyfpgrowth.find_frequent_patterns(transactions,support_threshold=2)
print('Frequent Patterns')
print(frequent)
```

```python
# Step 3: Generate Rules
rules = pyfpgrowth.generate_association_rules(frequent, confidence_threshold=0.5)
print('Association Rules')
print(rules)
```

```python
# 1.) Bread --> Egg : 0.5
# 2.) Egg --> bread : 0.6 both are greater than 50%   , meaning if we recommend to buy bread if cuat add egg in his cart [True]

# But the confidence value is different
# behind the scene there are 2 rules
  # Rule 1: of all the basket with Bread, how many also include Egg
  # RUle 2 : of all the baskets with Egg, how many also include Bread

# count how many people bought Bread  : Bread appear 4 times
# count how many people bought Egg     : Bread appear 3 times
# count how many people bought bread and egg : bread + egg 2 times
# confidence of bread bread + egg 2/4 = 0.5
# confidence of Egg + Bread = 0.666
```

```python
# simple Recommender
def recommend (bought_item,rules):
  for key in rules:
    if bought_item in key:
      recommended_item=rules[key][0]
      confidence = rules[key][1]
      return recommended_item,confidence
  return[],0

# test the recommendation
item,conf=recommend('Milk',rules)
print('Customer Who Bought Milk. Also Bought',item,conf)
print(f'{int(conf*100)}%')
```
