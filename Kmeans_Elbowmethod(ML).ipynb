```python
# import lobraries
import numpy as np
import matplotlib.pyplot as mb
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import warnings
warnings.filterwarnings('ignore')

# create a sample datasets or create synthetic cluster data
x,y = make_blobs(n_samples= 300,     # number of points
                 centers= 4,         # number of clusters
                 random_state= 42)
```

```python
#0--> column 0
# 1 --> column 1
# : --> show rows

mb.scatter(x[:,0],x[:,1],c=y)  # c=y is color
mb.title('Cluester generated with make_blobs')
mb.xlabel('Features 1')
mb.ylabel('Features 2')
mb.show()
```

```python
# Elbow method
# list to record WCS

wcss=[]
for k in range(1,11):     # test k= 1 to 10
    kmeans = KMeans(n_clusters=k,random_state=42)
    kmeans.fit(x)        # checking the best k for Data Points
    wcss.append(kmeans.inertia_)     #

mb.plot(range(1,11),wcss,marker='o')
mb.title('Elbow Method')
mb.xlabel('Number of Clusters')
mb.ylabel('WCSS')
mb.grid(True)
mb.show()
```

```python
# predict new point
new_point= [[-9.0772758 ,   6.41595925]]
prediction= kmeans.predict(new_point)
print("Predicted cluster for the new point is:", prediction)
```

```python
mb.figure(figsize=(8,6))
mb.scatter(x[:,0],x[:,1],c=y)
mb.scatter(new_point[0][0],new_point[0][1],c='red',marker='x',s=200)  # c=y is color
mb.title('New Point Clusters')
mb.xlabel('Features 1')
mb.ylabel('Features 2')
mb.grid(True)
mb.show()
```
